```python
df = pd.DataFrame(documents)
df
```

```python
df.columns = ["Titulo","Conteudo"]
df
```

```python
model="models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                             content=text,
                             title=title,
                             task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embedding"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Conteudo"]), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, database, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                              content=consulta,
                                              task_type="RETRIEVAL_QUERY")["embedding"]
  produtos_escalares = np.dot(np.stack(database["Embedding"]), embedding_da_consulta)
  indice = np.argmax(produtos_escalares)
  return database.iloc[indice]["Conteudo"]
```

```python
consulta = "How can I shift in google car?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```

```python
prompt = f"Rescreva esse texto de uma forma mais descontraida, sem adicionar informações que não façam parte do texto: {trecho}"

model_2 = genai.GenerativeModel("gemini-1.0-pro")
response = model_2.generate_content(prompt)
print(response.text)
```

```python
generation_config = {
    "temperature": 0.5,
    "candidate_count": 1,
}

prompt = f"Rescreva esse texto de uma forma mais descontraida, sem adicionar informações que não façam parte do texto: {trecho}"

model_2 = genai.GenerativeModel("gemini-1.0-pro", generation_config=generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
